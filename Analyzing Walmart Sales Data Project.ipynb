```python
##### Visualizations

## Distribution plots for numerical variables
numerical_columns = ['Unit price', 'Quantity', 'Total', 'Rating']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns):
    plt.subplot(2, 2, i+1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

## Count plots for categorical variables
categorical_columns = ['Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Payment']

plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_columns):
    plt.subplot(3, 2, i+1)
    sns.countplot(y=data[col], order=data[col].value_counts().index)
    plt.title(f'Count of {col}')

plt.tight_layout()
plt.show()

## Correlation heatmap
plt.figure(figsize=(10, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
##### Visualizations Summary:

## 1. Distribution Plots for Numerical Variables: Shows the distribution of Unit price, Quantity, Total, and Rating.
## 2. Count Plots for Categorical Variables: Shows the count of each category in Branch, City, Customer type, Gender, Product 
##    line, and Payment.
## 3. Correlation Heatmap: Displays the correlation between numerical variables, indicating relationships and potential 
##    multicollinearity.
```

```python
##### 4. Feature Engineering Analysis:
```

```python
# Convert Date and Time to datetime
data['Date'] = pd.to_datetime(data['Date'])
data['Time'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.time

# Extracting new date features
data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.hour

# Customer Lifetime Value
data['Customer Lifetime Value'] = data.groupby('Invoice ID')['Total'].transform('sum')

# Product Category Count
data['Product Category Count'] = data.groupby('Invoice ID')['Product line'].transform('nunique')

# Display first few rows with new features
data.head()
```

```python
##### Feature Engineering Summary:

## 1. Day: Extracted from Date.
## 2. Month: Extracted from Date.
## 3. Hour: Extracted from Time.
## 4. Customer Lifetime Value: Total spending by a customer (based on Invoice ID).
## 5. Product Category Count: Count of different product categories purchased by a customer (based on Invoice ID).
```

```python
##### 5. Other Analysis and Insights
```

```python
##### A. Sales Performance by Branch and City

## Grouping sales data by branch and city
sales_by_branch = data.groupby('Branch')['Total'].sum()
sales_by_city = data.groupby('City')['Total'].sum()

## Plotting total sales by branch
if not sales_by_branch.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sales_by_branch.index, y=sales_by_branch.values)
    plt.title('Total Sales by Branch')
    plt.xlabel('Branch')
    plt.ylabel('Total Sales')
    plt.show()
else:
    print("No data available for sales by branch.")

## Plotting sales distribution by city
if not sales_by_city.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sales_by_city.index, y=sales_by_city.values)
    plt.title('Sales Distribution by City')
    plt.xlabel('City')
    plt.ylabel('Total Sales')
    plt.show()
else:
    print("No data available for sales by city.")
```

```python
##### Summary of Sales Performance by Branch and City:

## 1. Group the data by 'Branch' and 'City' to calculate the total sales for each.
## 2. Use sns.barplot to create bar plots for visualizing total sales by branch and city.
## 3. Add checks to ensure the sequences are not empty before plotting.
```

```python
##### B. Customer Demographics

## Grouping sales data by customer type and gender
sales_by_customer_type = data.groupby('Customer type')['Total'].sum()
sales_by_gender = data.groupby('Gender')['Total'].sum()

## Plotting sales by customer type
if not sales_by_customer_type.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sales_by_customer_type.index, y=sales_by_customer_type.values)
    plt.title('Sales by Customer Type')
    plt.xlabel('Customer Type')
    plt.ylabel('Total Sales')
    plt.show()
else:
    print("No data available for sales by customer type.")

## Plotting gender-based sales analysis
if not sales_by_gender.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sales_by_gender.index, y=sales_by_gender.values)
    plt.title('Gender-Based Sales Analysis')
    plt.xlabel('Gender')
    plt.ylabel('Total Sales')
    plt.show()
else:
    print("No data available for sales by gender.")
```

```python
##### Summary of Customer Demographics:

## 1. Group the data by 'Customer type' and 'Gender' to calculate total sales.
## 2. Create bar plots to visualize sales by customer type and gender.
## 3. Add checks to ensure the sequences are not empty before plotting.
```

```python
##### C. Product Line Performance

## Grouping revenue and sales volume by product line
revenue_by_product_line = data.groupby('Product line')['Total'].sum()
sales_volume_by_product_line = data.groupby('Product line')['Quantity'].sum()

## Plotting revenue by product line
if not revenue_by_product_line.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=revenue_by_product_line.index, y=revenue_by_product_line.values)
    plt.title('Revenue by Product Line')
    plt.xlabel('Product Line')
    plt.ylabel('Total Revenue')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No data available for revenue by product line.")

## Plotting sales volume by product line
if not sales_volume_by_product_line.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=sales_volume_by_product_line.index, y=sales_volume_by_product_line.values)
    plt.title('Sales Volume by Product Line')
    plt.xlabel('Product Line')
    plt.ylabel('Total Quantity Sold')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No data available for sales volume by product line.")
```

```python
##### Summary of Product Line Performance:

## 1. Group the data by 'Product line' to calculate total revenue and sales volume (quantity sold) for each product line.
## 2. Use bar plots to visualize revenue and sales volume by product line.
## 3. Add checks to ensure the sequences are not empty before plotting.
```

```python
##### D. Pricing and Quantity

## Scatter plot of unit price vs. quantity sold
if not data.empty:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=data['Unit price'], y=data['Quantity'])
    plt.title('Unit Price vs. Quantity')
    plt.xlabel('Unit Price')
    plt.ylabel('Quantity Sold')
    plt.show()
else:
    print("No data available for unit price vs. quantity.")

## Heatmap of total sales based on unit price and quantity
total_sales_analysis = data.groupby(['Unit price', 'Quantity'])['Total'].sum().reset_index()
if not total_sales_analysis.empty:
    plt.figure(figsize=(10, 6))
    sales_pivot = total_sales_analysis.pivot('Unit price', 'Quantity', 'Total')
    sns.heatmap(sales_pivot, cmap='viridis')
    plt.title('Total Sales Analysis')
    plt.xlabel('Quantity')
    plt.ylabel('Unit Price')
    plt.show()
else:
    print("No data available for total sales analysis.")
```

```python
##### Summary of Pricing and Quantity:

## 1. Create a scatter plot to explore the relationship between unit price and quantity sold.
## 2. Generate a heatmap to analyze total sales based on unit price and quantity.
## 3. Add checks to ensure the sequences are not empty before plotting.
```

```python
##### E. Payment Methods

## Counting transactions and grouping sales by payment method
payment_method_counts = data['Payment'].value_counts()
payment_method_sales = data.groupby('Payment')['Total'].sum()

## Plotting popular payment methods
if not payment_method_counts.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=payment_method_counts.index, y=payment_method_counts.values)
    plt.title('Popular Payment Methods')
    plt.xlabel('Payment Method')
    plt.ylabel('Number of Transactions')
    plt.show()
else:
    print("No data available for payment methods.")

## Plotting payment method impact on sales
if not payment_method_sales.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=payment_method_sales.index, y=payment_method_sales.values)
    plt.title('Payment Method Impact on Sales')
    plt.xlabel('Payment Method')
    plt.ylabel('Total Sales')
    plt.show()
else:
    print("No data available for payment method impact on sales.")
```

```python
##### Summary of Payment Methods:

## 1. Count the number of transactions for each payment method and group sales by payment method.
## 2. Create bar plots to visualize the popularity of payment methods and their impact on sales.
## 3. Add checks to ensure the sequences are not empty before plotting.
```

```python
##### F. Sales Trends Over Time

## Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

## Extract day, month, year, and hour
data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M:%S').dt.hour

## Daily Sales
daily_sales = data.groupby('Date')['Total'].sum()

## Monthly Sales
monthly_sales = data.groupby(['Year', 'Month'])['Total'].sum()

## Hourly Sales
hourly_sales = data.groupby('Hour')['Total'].sum()

## Plotting
plt.figure(figsize=(18, 6))

## Check and Plot Daily Sales
if not daily_sales.empty:
    plt.subplot(1, 3, 1)
    daily_sales.plot()
    plt.title('Daily Sales Trend')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
else:
    print("No data available for Daily Sales Trend.")

## Check and Plot Monthly Sales
if not monthly_sales.empty:
    plt.subplot(1, 3, 2)
    monthly_sales.plot(kind='bar')
    plt.title('Monthly Sales Trend')
    plt.xlabel('Month-Year')
    plt.ylabel('Total Sales')
else:
    print("No data available for Monthly Sales Trend.")

## Check and Plot Hourly Sales
if not hourly_sales.empty:
    plt.subplot(1, 3, 3)
    hourly_sales.plot(kind='bar')
    plt.title('Hourly Sales Trend')
    plt.xlabel('Hour')
    plt.ylabel('Total Sales')
else:
    print("No data available for Hourly Sales Trend.")

plt.tight_layout()
plt.show()
```

```python
##### Summary of Sales Trends Over Time:

## 1. Data Conversion: Ensure the Date and Time columns are in the correct format for analysis.
## 2. Data Aggregation: Sum the Total sales for each time unit (day, month, hour).
## 3. Empty Sequence Checks: Before plotting, check if the aggregated data is non-empty to avoid errors.
```

```python
##### G. Customer Feedback

## Counting ratings and identifying low-rating product lines
rating_distribution = data['Rating'].value_counts()
low_ratings = data[data['Rating'] <= 3]
low_rating_product_lines = low_ratings['Product line'].value_counts()

## Plotting rating distribution
if not rating_distribution.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=rating_distribution.index, y=rating_distribution.values)
    plt.title('Rating Distribution')
    plt.xlabel('Rating')
    plt.ylabel('Number of Ratings')
    plt.show()
else:
    print("No data available for rating distribution.")

## Plotting areas for improvement based on low ratings
if not low_rating_product_lines.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=low_rating_product_lines.index, y=low_rating_product_lines.values)
    plt.title('Areas for Improvement')
    plt.xlabel('Product Line')
    plt.ylabel('Number of Low Ratings')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No data available for areas for improvement.")
```

```python
##### Customer Feedback:

## 1. Count the number of ratings for each rating value.
## 2. Identify product lines associated with low ratings (ratings <= 3).
## 3. Create bar plots to visualize rating distribution and areas for improvement based on low ratings.
## 4. Add checks to ensure the sequences are not empty before plotting.
```
